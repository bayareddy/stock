```python
import os
import sys
sys.path.insert(0, r'D:\\bayareddy\\stock')
import pandas as pd
import datetime as dt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas_ta as ta
import fyers_util as fu

def getStockData(stock_name, time_frame, range_from, range_to):
    """
    Fetches historical stock data in chunks of 100 days (or bars) and 
    returns a final DataFrame with a datetime index.
    """
    try:
        cur_start = range_from
        all_candles = [] # List to store raw candle data (list of lists)
        columns = ['timestamp', 'open', 'high', 'low', 'close', 'volume']
        
        # --- Data Fetching Loop to handle 100-day chunks ---
        while cur_start <= range_to:
            # Determine the end date for the current 100-day chunk
            # We use 99 days added to 'cur_start' to make a 100-day range (start day + 99 days)
            cur_end = min(cur_start + dt.timedelta(days=99), range_to) 
            
            # Call the API for the current chunk
            response = fu.history(stock_name, time_frame, cur_start, cur_end)
            
            if response.get('s') == 'ok' and response.get('candles'):
                # Append the list of candles (data) from the response to our main list
                all_candles.extend(response.get('candles'))
            else:
                # Handle error or end of data for the current chunk
                error_msg = response.get('message', 'Unknown API error')
                if response.get('s') != 'ok':
                    print(f"Warning: API error on chunk {cur_start} to {cur_end}: {error_msg}")
                    # If an error occurs, we stop fetching to avoid infinite loops or repeated errors
                    break
            
            # Move to the start date for the next chunk (1 day after the current end date)
            cur_start = cur_end + dt.timedelta(days=1)
            
        # --- DataFrame Creation and Formatting ---
        if not all_candles:
            print("No data received from API after chunking.")
            return pd.DataFrame()

        # 1. Create the DataFrame in a single step from the collected list of lists
        df = pd.DataFrame(all_candles, columns=columns) 
        
        # 2. Convert timestamp (Unix time) to datetime objects
        df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
        
        # 3. Set the datetime column as the index
        df.set_index('timestamp', inplace=True)
        
        # 4. Ensure chronological order (Oldest to Newest) and remove duplicates if API returns any overlap
        df = df.sort_index(ascending=True).drop_duplicates()
        
        return df
        
    except Exception as e:
        print(f"An exception occurred: {e}")
        return None

def plot_gapless_candlestick(df, stock_name):
# ... (plot_gapless_candlestick function remains unchanged) ...
    full_date_range = pd.date_range(start=df.index.min(), end=df.index.max())
    missing_dates = full_date_range.difference(df.index)

    fig = make_subplots( rows=3, cols=1, shared_xaxes=True, vertical_spacing=0.03,
        subplot_titles=(f'{stock_name} Candlesticks', 'Volume', 'RSI (14)'),
        row_heights=[0.6, 0.2, 0.2] )

    # Candlestick Trace (Row 1)
    fig.add_trace(go.Candlestick(x=df.index, open=df['open'], high=df['high'], 
                                 low=df['low'], close=df['close'], name='OHLC'), row=1, col=1)
    # Volume Trace (Row 2)
    fig.add_trace(go.Bar(x=df.index, y=df['volume'], name='Volume', 
                          marker_color='rgba(0, 100, 250, 0.3)'), row=2, col=1)
    # RSI Trace (Row 3)
    fig.add_trace(go.Scatter(x=df.index, y=df['RSI_14'], name='RSI',
                              line=dict(color='yellow', width=1)), row=3, col=1)
    # RSI Lines
    fig.add_hline(y=70, line_dash="dash", line_color="red", line_width=1, row=3, col=1)
    fig.add_hline(y=30, line_dash="dash", line_color="green", line_width=1, row=3, col=1)
    # Style
    fig.update_layout( title=f'{stock_name} - Last 30 Days (Gapless with RSI)',
        yaxis_title='Price (INR)', yaxis2_title='Volume', yaxis3_title='RSI',        
        template='plotly_dark', xaxis_rangeslider_visible=False, showlegend=False)
    # Gapless
    fig.update_xaxes(rangebreaks=[dict(values=missing_dates)])
    print("Displaying interactive chart with full RSI...")
    fig.show()

# --- Main execution block ---
if __name__ == "__main__":
    stock_to_fetch = "NSE:NIFTY50-INDEX"
    time_frame = 'D' 
    today = dt.date.today()
    days_to_display = 60
    rsi_period = 14
    
    # Fetch enough data for calculation buffer (e.g., 60 days to display + 90 buffer days = 150 days)
    buffer_days = 120 # Enough buffer for RSI calculation
    start_date_for_fetch = today - dt.timedelta(days=days_to_display + buffer_days)
    
    print(f"Fetching data from {start_date_for_fetch} to {today} in 100-day chunks...")
    
    # 1. Fetch the (longer) full dataset
    stock_df_full = getStockData( stock_name=stock_to_fetch, time_frame=time_frame, range_from=start_date_for_fetch, 
        range_to=today)
    
    if stock_df_full is not None and not stock_df_full.empty:
        print(f"Successfully fetched {len(stock_df_full)} data points.")
        
        # 2. Calculate RSI on the *full* dataset
        stock_df_full['RSI_14'] = ta.rsi(stock_df_full['close'], length=rsi_period)
        
        # 3. Trim the DataFrame to only the last 'days_to_display' trading days
        stock_df_display = stock_df_full.iloc[-days_to_display:]

        print(f"Plotting the last {len(stock_df_display)} days of data...")
        
        # 4. Plot the trimmed DataFrame
        plot_gapless_candlestick(stock_df_display, stock_to_fetch)
    else:
        print("Failed to fetch or process stock data. Check token and parameters.")
```
